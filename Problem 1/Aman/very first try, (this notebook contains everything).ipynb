# Outlier determination

```python
def count_outliers(data, column_name):
    column_data = data[column_name]
    # Calculate Q1 and Q3
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    # Calculate IQR
    IQR = Q3 - Q1
    # Min and Max Calculation
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Count outliers
    outliers = column_data[(column_data<lower_bound) | (column_data>upper_bound)]
    return len(outliers)
```

```python
attributes = ['MEAT', 'UNIT_PRICE', 'MILK', 'BOX_ID']

for attribute in attributes:
    count = count_outliers(merged_df,attribute)
    print(attribute,'- ',count,'out of',len(merged_df[attribute]))
```

# Making the TARGET COLUMN to predict


```python
# Custom function to map the PURCHASE column based on the condition
def map_purchase(day):
    if day >= 1 and day <= 15:
        return 'Y'
    else:
        return 'N'

# Apply the custom function to map the PURCHASE column
merged_df['PURCHASE'] = merged_df['DAY'].apply(map_purchase)
```
