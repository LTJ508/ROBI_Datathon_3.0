```python
df_customer = pd.read_csv('Datasets/df_customer_train.csv')
```

```python
df_customer.info()
```

```python
df_customer['LAST_PURCHASE_DATE'] = pd.to_datetime(df_customer['LAST_PURCHASE_DATE'])
df_customer['DAY_SINCE_LAST_PURCHASE'] = pd.to_timedelta(df_customer['DAY_SINCE_LAST_PURCHASE'])
df_customer['AVERAGE_DIFF'] = pd.to_timedelta(df_customer['AVERAGE_DIFF'])
```

```python
df_customer.info()
```

```python
df_customer.describe()
```

```python
october_2018_last_15 = df.groupby('MAGIC_KEY')['PURCHASE_DATE'].apply(
    lambda x: 1 if ((x.dt.month == 10) & (x.dt.day > 15) & (x.dt.year == 2018)).any() else 0
).reset_index(name='10_2018_LAST_15')

# Display the result DataFrame
october_2018_last_15
```

```python
df_customer = pd.merge(df_customer, october_2018_last_15, on='MAGIC_KEY')
df_customer
```

```python
november_2018_last_15 = df.groupby('MAGIC_KEY')['PURCHASE_DATE'].apply(
    lambda x: 1 if ((x.dt.month == 11) & (x.dt.day > 15) & (x.dt.year == 2018)).any() else 0
).reset_index(name='11_2018_LAST_15')

# Display the result DataFrame
november_2018_last_15
```

```python
df_customer = pd.merge(df_customer, november_2018_last_15, on='MAGIC_KEY')
df_customer
```

```python
december_2018_last_15 = df.groupby('MAGIC_KEY')['PURCHASE_DATE'].apply(
    lambda x: 1 if ((x.dt.month == 12) & (x.dt.day > 15) & (x.dt.year == 2018)).any() else 0
).reset_index(name='12_2018_LAST_15')

# Display the result DataFrame
december_2018_last_15
```

```python
df_customer = pd.merge(df_customer, december_2018_last_15, on='MAGIC_KEY')
df_customer
```

```python
january_2018_last_15 = df.groupby('MAGIC_KEY')['PURCHASE_DATE'].apply(
    lambda x: 1 if ((x.dt.month == 1) & (x.dt.day > 15) & (x.dt.year == 2019)).any() else 0
).reset_index(name='1_2019_LAST_15')

# Display the result DataFrame
january_2018_last_15
```

```python
df_customer = pd.merge(df_customer, january_2018_last_15, on='MAGIC_KEY')
df_customer
```
