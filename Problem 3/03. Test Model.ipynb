```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Datasets/train.csv')
```

```python
df
```

```python
df.info()
```

```python
df['LAST_PURCHASE_DATE'] = pd.to_datetime(df['LAST_PURCHASE_DATE'])
df['DAY_SINCE_LAST_PURCHASE'] = pd.to_timedelta(df['DAY_SINCE_LAST_PURCHASE'])
df['AVERAGE_DIFF'] = pd.to_timedelta(df['AVERAGE_DIFF'])
```

```python
df['DAY_SINCE_LAST_PURCHASE'] = df['DAY_SINCE_LAST_PURCHASE'].dt.days.astype(int)
df['AVERAGE_DIFF'] = df['AVERAGE_DIFF'].dt.days.astype(int)
```

```python
df.drop(columns=['LAST_PURCHASE_DATE'], inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder #importing the labenconder
le = LabelEncoder() #creating an instance of LabelEncoder

df['MAGIC_KEY_ENCODED'] = le.fit_transform(df['MAGIC_KEY'])
df_encoded = df.drop(columns = ['MAGIC_KEY'])
```

```python
df_encoded.head()
```

```python
correlation = df_encoded.corr(method = 'pearson') #finding all correlation between each features
print(correlation['MEAT_1_2019_FIRST_15'].sort_values(ascending = False).to_string()) #printing correlation with class in descending order
```

```python
from sklearn.utils import shuffle #importing the suffle module
shuffled_df = shuffle(df_encoded)
shuffled_df #checking
```

```python
#rearranging the index value
rearranged_df = shuffled_df.reset_index(drop = True)
rearranged_df
```

```python
X = rearranged_df.drop(columns = ['MEAT_1_2019_FIRST_15']) #dropping the class collumn to separate the feature
X
```

```python
y =  rearranged_df['MEAT_1_2019_FIRST_15'] #separating target to y
y
```

```python
from sklearn.model_selection import train_test_split #importing module
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20) #80-20% split
```

```python
print(len(X_train)) #checking
print(len(X_test))
```

```python
print(len(y_train)) #checking
print(len(y_test))
```

#using MinMax Scaler
from sklearn.preprocessing import MinMaxScaler #importing scaler

scaler = MinMaxScaler() #creating an instance

#we can scale the features only, not the target
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

# <font color="red">Linear Regression</font>

```python
#importing the model first
from sklearn.linear_model import LinearRegression
reg = LinearRegression() #creating an instance of the model using default parameter
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import math
```

```python
R2 = r2_score(y_pred, y_test) #checking R2 Score
print("R2: ", R2)

MSE = mean_squared_error(y_pred, y_test) #checking Mean Squared Error
print("MSE: ", MSE)

RMSE = math.sqrt(MSE) #checking Root Mean Squared Error
print("RMSE: ", RMSE)

MAE = mean_absolute_error(y_pred, y_test) #cehcking Mean Absolute Error
print("MAE: ", MAE)
```

# <font color="red"> Random Forest</font>

```python
from sklearn.ensemble import RandomForestRegressor
rand_forest = RandomForestRegressor(n_estimators = 100, random_state = 42)
rand_forest.fit(X_train, y_train)
```

```python
y_pred = rand_forest.predict(X_test)
```

```python
R2 = r2_score(y_pred, y_test) #checking R2 Score
print("R2: ", R2)

MSE = mean_squared_error(y_pred, y_test) #checking Mean Squared Error
print("MSE: ", MSE)

RMSE = math.sqrt(MSE) #checking Root Mean Squared Error
print("RMSE: ", RMSE)

MAE = mean_absolute_error(y_pred, y_test) #cehcking Mean Absolute Error
print("MAE: ", MAE)
```

# <font color="red">SVR</font>

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
svr_regressor = SVR(kernel='rbf', C=100, gamma=0.1)

# Train the SVR
svr_regressor.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = svr_regressor.predict(X_test)
```

```python
R2 = r2_score(y_pred, y_test) #checking R2 Score
print("R2: ", R2)

MSE = mean_squared_error(y_pred, y_test) #checking Mean Squared Error
print("MSE: ", MSE)

RMSE = math.sqrt(MSE) #checking Root Mean Squared Error
print("RMSE: ", RMSE)

MAE = mean_absolute_error(y_pred, y_test) #cehcking Mean Absolute Error
print("MAE: ", MAE)
```

# <font color="red">Neural Net</font>

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

```python
# Define the neural network model
model = Sequential()
model.add(Input(shape=(X_train.shape[1],)))  # Input layer
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))  # Output layer with linear activation for regression
```

```python
# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)
```

```python
# Evaluate the model on test data
mse = model.evaluate(X_test, y_test, verbose=0)
print('Mean Squared Error on test data:', mse)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
R2 = r2_score(y_pred, y_test) #checking R2 Score
print("R2: ", R2)

MSE = mean_squared_error(y_pred, y_test) #checking Mean Squared Error
print("MSE: ", MSE)

RMSE = math.sqrt(MSE) #checking Root Mean Squared Error
print("RMSE: ", RMSE)

MAE = mean_absolute_error(y_pred, y_test) #cehcking Mean Absolute Error
print("MAE: ", MAE)
```
