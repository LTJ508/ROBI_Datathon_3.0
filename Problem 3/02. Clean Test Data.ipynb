```python
import pandas as pd
```

```python
df = pd.read_csv('../Common Datasets/test_df.csv')
df
```

```python
df['PURCHASE_DATE'] = pd.to_datetime(df['PURCHASE_DATE'])
```

```python
df.info()
```

```python
df = df[df['PURCHASE_DATE'].dt.day <= 15]
df = df[df['BOX_COUNT'] >= 1]
df
```

```python
df['MAGIC_KEY'].nunique()
```

```python
df.info()
```

```python
test_df = df.groupby('MAGIC_KEY')['MEAT'].sum().reset_index()
test_df['MEAT'] = test_df['MEAT'].fillna(0)
test_df
```

```python
df_train = pd.read_csv('../Common Datasets/train_df.csv')
df_train
```

```python
train_df = df_train.groupby('MAGIC_KEY')['MEAT'].sum().reset_index()
train_df['MEAT'] = train_df['MEAT'].fillna(0)
train_df
```

```python
merged_df = pd.merge(test_df, train_df, on='MAGIC_KEY', how='inner')
```

```python
len(merged_df)
```
