```python
boxes = pd.read_csv('../Common Datasets/boxes.csv')
boxes
```

```python
df_customer = pd.merge(df_customer, boxes, on='BOX_ID', how='left')
df_customer
```

```python
df_customer = df_customer.drop(columns = ['QUALITY'])
df_customer = df_customer.drop(columns = ['DELIVERY_OPTION'])
df_customer.rename(columns={'BOX_ID': 'MODE_BOX_ID'}, inplace=True)
df_customer.rename(columns={'MILK': 'MODE_BOX_ID_MILK'}, inplace=True)
df_customer.rename(columns={'MEAT': 'MODE_BOX_ID_MEAT'}, inplace=True)
df_customer.rename(columns={'UNIT_PRICE': 'MODE_BOX_ID_UNIT_PRICE'}, inplace=True)

df_customer
```

```python
# Filter the DataFrame to include only dates in the first 15 days of October 2018
october_first_15_df = df[(df['PURCHASE_DATE'].dt.year == 2018) & 
                          (df['PURCHASE_DATE'].dt.month == 10) &
                          (df['PURCHASE_DATE'].dt.day <= 15)]

november_first_15_df = df[(df['PURCHASE_DATE'].dt.year == 2018) & 
                          (df['PURCHASE_DATE'].dt.month == 11) &
                          (df['PURCHASE_DATE'].dt.day <= 15)]

december_first_15_df = df[(df['PURCHASE_DATE'].dt.year == 2018) & 
                          (df['PURCHASE_DATE'].dt.month == 12) &
                          (df['PURCHASE_DATE'].dt.day <= 15)]

january_first_15_df = df[(df['PURCHASE_DATE'].dt.year == 2019) & 
                          (df['PURCHASE_DATE'].dt.month == 1) &
                          (df['PURCHASE_DATE'].dt.day <= 15)]
```

```python
milk_october_first_15 = october_first_15_df.groupby('MAGIC_KEY')['MILK'].sum().reset_index()
milk_october_first_15.rename(columns={'MILK': 'MILK_10_2018_FIRST_15'}, inplace=True)

meat_october_first_15 = october_first_15_df.groupby('MAGIC_KEY')['MEAT'].sum().reset_index()
meat_october_first_15.rename(columns={'MEAT': 'MEAT_10_2018_FIRST_15'}, inplace=True)


milk_november_first_15 = november_first_15_df.groupby('MAGIC_KEY')['MILK'].sum().reset_index()
milk_november_first_15.rename(columns={'MILK': 'MILK_11_2018_FIRST_15'}, inplace=True)

meat_november_first_15 = november_first_15_df.groupby('MAGIC_KEY')['MEAT'].sum().reset_index()
meat_november_first_15.rename(columns={'MEAT': 'MEAT_11_2018_FIRST_15'}, inplace=True)


milk_december_first_15 = december_first_15_df.groupby('MAGIC_KEY')['MILK'].sum().reset_index()
milk_december_first_15.rename(columns={'MILK': 'MILK_12_2018_FIRST_15'}, inplace=True)

meat_december_first_15 = december_first_15_df.groupby('MAGIC_KEY')['MEAT'].sum().reset_index()
meat_december_first_15.rename(columns={'MEAT': 'MEAT_12_2018_FIRST_15'}, inplace=True)


#milk_january_first_15 = january_first_15_df.groupby('MAGIC_KEY')['MILK'].sum().reset_index()
#milk_january_first_15.rename(columns={'MILK': 'MILK_1_2019_FIRST_15'}, inplace=True)

meat_january_first_15 = january_first_15_df.groupby('MAGIC_KEY')['MEAT'].sum().reset_index()
meat_january_first_15.rename(columns={'MEAT': 'MEAT_1_2019_FIRST_15'}, inplace=True)


df_customer = pd.merge(df_customer, milk_october_first_15, on='MAGIC_KEY', how='left')
df_customer = pd.merge(df_customer, meat_october_first_15, on='MAGIC_KEY', how='left')

df_customer = pd.merge(df_customer, milk_november_first_15, on='MAGIC_KEY', how='left')
df_customer = pd.merge(df_customer, meat_november_first_15, on='MAGIC_KEY', how='left')

df_customer = pd.merge(df_customer, milk_december_first_15, on='MAGIC_KEY', how='left')
df_customer = pd.merge(df_customer, meat_december_first_15, on='MAGIC_KEY', how='left')

#df_customer = pd.merge(df_customer, milk_january_first_15, on='MAGIC_KEY', how='left')
df_customer = pd.merge(df_customer, meat_january_first_15, on='MAGIC_KEY', how='left')


df_customer['MILK_10_2018_FIRST_15'] = df_customer['MILK_10_2018_FIRST_15'].fillna(0)
df_customer['MEAT_10_2018_FIRST_15'] = df_customer['MEAT_10_2018_FIRST_15'].fillna(0)

df_customer['MILK_11_2018_FIRST_15'] = df_customer['MILK_11_2018_FIRST_15'].fillna(0)
df_customer['MEAT_11_2018_FIRST_15'] = df_customer['MEAT_11_2018_FIRST_15'].fillna(0)

df_customer['MILK_12_2018_FIRST_15'] = df_customer['MILK_12_2018_FIRST_15'].fillna(0)
df_customer['MEAT_12_2018_FIRST_15'] = df_customer['MEAT_12_2018_FIRST_15'].fillna(0)

#df_customer['MILK_1_2019_FIRST_15'] = df_customer['MILK_1_2019_FIRST_15'].fillna(0)
df_customer['MEAT_1_2019_FIRST_15'] = df_customer['MEAT_1_2019_FIRST_15'].fillna(0)


df_customer
```

```python
# Group by 'MAGIC_KEY' and sum the 'PURCHASE_DATE' column for each group
total_purchase_october_first_15 = october_first_15_df.groupby('MAGIC_KEY')['PURCHASE_DATE'].count().reset_index()
total_purchase_october_first_15.columns = ['MAGIC_KEY', 'T_PURCHASE_10_2018_FIRST_15']

# Group by 'MAGIC_KEY' and sum the 'PURCHASE_DATE' column for each group
total_purchase_november_first_15 = november_first_15_df.groupby('MAGIC_KEY')['PURCHASE_DATE'].count().reset_index()
total_purchase_november_first_15.columns = ['MAGIC_KEY', 'T_PURCHASE_11_2018_FIRST_15']

# Group by 'MAGIC_KEY' and sum the 'PURCHASE_DATE' column for each group
total_purchase_december_first_15 = december_first_15_df.groupby('MAGIC_KEY')['PURCHASE_DATE'].count().reset_index()
total_purchase_december_first_15.columns = ['MAGIC_KEY', 'T_PURCHASE_12_2018_FIRST_15']

# Group by 'MAGIC_KEY' and sum the 'PURCHASE_DATE' column for each group
total_purchase_january_first_15 = january_first_15_df.groupby('MAGIC_KEY')['PURCHASE_DATE'].count().reset_index()
total_purchase_january_first_15.columns = ['MAGIC_KEY', 'T_PURCHASE_1_2019_FIRST_15']


df_customer = pd.merge(df_customer, total_purchase_october_first_15, on='MAGIC_KEY', how='left')
df_customer = pd.merge(df_customer, total_purchase_november_first_15, on='MAGIC_KEY', how='left')
df_customer = pd.merge(df_customer, total_purchase_december_first_15, on='MAGIC_KEY', how='left')
df_customer = pd.merge(df_customer, total_purchase_january_first_15, on='MAGIC_KEY', how='left')

df_customer['T_PURCHASE_10_2018_FIRST_15'] = df_customer['T_PURCHASE_10_2018_FIRST_15'].fillna(0)
df_customer['T_PURCHASE_11_2018_FIRST_15'] = df_customer['T_PURCHASE_11_2018_FIRST_15'].fillna(0)
df_customer['T_PURCHASE_12_2018_FIRST_15'] = df_customer['T_PURCHASE_12_2018_FIRST_15'].fillna(0)
df_customer['T_PURCHASE_1_2019_FIRST_15'] = df_customer['T_PURCHASE_1_2019_FIRST_15'].fillna(0)

df_customer
```
