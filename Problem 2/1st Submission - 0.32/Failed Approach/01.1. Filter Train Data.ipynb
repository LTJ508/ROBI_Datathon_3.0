```python
import pandas as pd
```

```python
df = pd.read_csv('../../Common Datasets/train_df.csv')
df
```

```python
df = df.drop(columns = ['PURCHASE'])
df['PURCHASE_DATE'] = pd.to_datetime(df['PURCHASE_DATE'])
```

```python
df.info()
```

```python
df = df[df['PURCHASE_DATE'].dt.day <= 15]
df = df[df['BOX_COUNT'] <= 1]
df
```

```python
df['MAGIC_KEY'].nunique()
```

```python
duplicates =  df[df['MAGIC_KEY'].duplicated(keep='first')] #it will return the duplicated row from the datasets
duplicates
```

```python
filtered_df = df[~df['MAGIC_KEY'].isin(duplicates['MAGIC_KEY'])]
filtered_df
```

```python
filtered_df['MAGIC_KEY'].nunique()
```

```python
filtered_df = filtered_df.drop(columns = ['PURCHASE_DATE'])
```

```python
filtered_df.to_csv("train2.csv", index=False)  # Save test data
```
