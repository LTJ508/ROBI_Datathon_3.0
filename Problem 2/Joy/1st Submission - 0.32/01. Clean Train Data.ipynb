```python
import pandas as pd
```

```python
df = pd.read_csv('Datasets/train_df.csv')
df
```

```python
df = df.drop(columns = ['PURCHASE'])
df = df.drop(columns = ['PURCHASE_DATE'])
```

```python
df.info()
```

```python
# Group by 'MAGIC_KEY' and calculate mode of 'BOX_ID' and mean of 'BOX_COUNT'
new_df = df.groupby('MAGIC_KEY').agg({'BOX_ID': lambda x: x.mode().iloc[0], 'BOX_COUNT': 'mean'}).reset_index()

# Convert 'BOX_COUNT' mean to integer
new_df['BOX_COUNT'] = new_df['BOX_COUNT'].astype(int)

# Display the new DataFrame
print(new_df)
```

```python
new_df
```

```python
new_df.describe()
```

```python
boxes = pd.read_csv('Datasets/boxes.csv')
boxes
```

```python
merged_df = pd.merge(boxes, new_df, on = 'BOX_ID', how = 'inner')
merged_df
```

```python
merged_df.to_csv("train.csv", index=False)  # Save test data
```
