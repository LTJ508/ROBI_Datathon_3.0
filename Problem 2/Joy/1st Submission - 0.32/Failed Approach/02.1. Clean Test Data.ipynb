```python
import pandas as pd
```

```python
df = pd.read_csv('../../Common Datasets/test_df.csv')
df
```

```python
df = df.drop(columns = ['PURCHASE'])
df['PURCHASE_DATE'] = pd.to_datetime(df['PURCHASE_DATE'])
```

```python
df.info()
```

```python
df = df[df['PURCHASE_DATE'].dt.day <= 15]
df = df[df['BOX_COUNT'] <= 1]
df
```

```python
df['MAGIC_KEY'].nunique()
```

```python
df.info()
```

```python
duplicates =  df[df['MAGIC_KEY'].duplicated(keep='first')] #it will return the duplicated row from the datasets
duplicates
```

```python
filtered_df = df[~df['MAGIC_KEY'].isin(duplicates['MAGIC_KEY'])]
filtered_df
```

```python
filtered_df['MAGIC_KEY'].nunique()
```

```python
test_df = filtered_df[['MAGIC_KEY', 'BOX_ID']].copy()
test_df
```

```python
df_train = pd.read_csv('../Datasets/train2.csv')
df_train
```

```python
train_df = df_train[['MAGIC_KEY', 'BOX_ID']].copy()
train_df
```

```python
merged_df = pd.merge(test_df, train_df, on='MAGIC_KEY', how='inner')
```
